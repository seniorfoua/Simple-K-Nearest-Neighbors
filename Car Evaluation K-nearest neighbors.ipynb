```python
import pandas as pd 
import numpy as np
import sklearn
import pickle
from sklearn.utils import shuffle
from sklearn import linear_model, preprocessing
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("car.data", names=["buying","maint","doors","persons","lug_boot","safety","class"])
data.head()
```

```python
predict = "class"
le = preprocessing.LabelEncoder()
buying = le.fit_transform(data["buying"])
maint = le.fit_transform(data["maint"])
doors = le.fit_transform(data["persons"])
lug_boot = le.fit_transform(data["lug_boot"])
safety = le.fit_transform(data["safety"])
clss = le.fit_transform(data["class"])
```

```python
x = list(zip(buying,maint,doors,lug_boot,safety))
y = list(clss)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1)
```

```python
best = 0
while best < 0.95 :
       x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1)
       model = KNeighborsClassifier(n_neighbors = 9)
       model = model.fit(x_train, y_train)
       acc = model.score(x_test,y_test)
       print ("accuracy: ", acc)
       if acc > best :
           best = acc
           with open ("carevaluation.pickle","wb") as f:
                 pickle.dump(model,f)
                
pickle_in = open("carevaluation.pickle","rb")
model = pickle.load(pickle_in)
```

```python
predicted = model.predict(x_test)
names = ["unacc","acc","good","vgood"]
for x in range(len(predicted)):
    print ("predicted: ",names[predicted[x]], "data: ", x_test[x], "actual: ", names[y_test[x]])
    n = model.kneighbors([x_test[x]],9,True)
    print("N: ", n)
```
